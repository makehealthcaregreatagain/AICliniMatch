```python
import requests, gzip, io, xml.etree.ElementTree as ET, csv, time, sys
from typing import Dict, Iterable, List, Optional

BASELINE_PREFIX = "25"  # "25" for 2025 baseline; change to "24" for 2024, etc.
BASE_URL = "https://ftp.ncbi.nlm.nih.gov/pubmed/baseline"

# CSV columns (fixed order so DictWriter is stable)
CSV_COLUMNS = [
    "pmid","date_published","date_revised","journal","title","authors",
    "key_substances","keywords","mesh_descriptors","mesh_descriptors_ui",
    "mesh_major_descriptors","mesh_qualifiers","mesh_qualifiers_ui","mesh_pairs","doi"
]

MONTH_MAP = {"Jan":"01","Feb":"02","Mar":"03","Apr":"04","May":"05","Jun":"06",
             "Jul":"07","Aug":"08","Sep":"09","Oct":"10","Nov":"11","Dec":"12"}

def norm_ymd(year: Optional[str], month: Optional[str], day: Optional[str]) -> Optional[str]:
    y = (year or "").strip(); m = (month or "").strip(); d = (day or "").strip()
    if not y: return None
    if m: m = MONTH_MAP.get(m, m); m = m.zfill(2)
    if d: d = d.zfill(2)
    if y and m and d: return f"{y}-{m}-{d}"
    if y and m: return f"{y}-{m}"
    return y

def find_first(elem, path) -> Optional[str]:
    x = elem.find(path)
    return x.text.strip() if (x is not None and x.text) else None

def get_pub_date(article) -> Optional[str]:
    j_pub = article.find(".//Journal/JournalIssue/PubDate")
    if j_pub is not None:
        s = norm_ymd(find_first(j_pub,"Year"), find_first(j_pub,"Month"), find_first(j_pub,"Day"))
        if s: return s
    for h in article.findall(".//PubmedData/History/PubMedPubDate"):
        if h.get("PubStatus") == "pubmed":
            s = norm_ymd(find_first(h,"Year"), find_first(h,"Month"), find_first(h,"Day"))
            if s: return s
    return None

def get_date_revised(article) -> Optional[str]:
    rev = article.find(".//MedlineCitation/DateRevised")
    return norm_ymd(find_first(rev,"Year"), find_first(rev,"Month"), find_first(rev,"Day")) if rev is not None else None

def get_journal(article) -> Optional[str]:
    return find_first(article, ".//Journal/Title") or find_first(article, ".//MedlineJournalInfo/MedlineTA")

def get_title(article) -> Optional[str]:
    return find_first(article, ".//Article/ArticleTitle")

def get_authors(article) -> Optional[str]:
    out = []
    for a in article.findall(".//AuthorList/Author"):
        coll = find_first(a, "CollectiveName")
        if coll: out.append(coll); continue
        last = find_first(a, "LastName"); fore = find_first(a, "ForeName") or find_first(a, "Initials")
        if last and fore: out.append(f"{last}, {fore}")
        elif last: out.append(last)
        elif fore: out.append(fore)
    return "; ".join(out) if out else None

def get_key_substances(article) -> Optional[str]:
    names = [find_first(c, "NameOfSubstance") for c in article.findall(".//ChemicalList/Chemical")]
    names = [n for n in names if n]
    return "; ".join(names) if names else None

def get_mesh(article) -> Dict[str, Optional[str]]:
    descriptors, descriptors_ui, majors = [], [], []
    qualifiers, qualifiers_ui, pairs = [], [], []

    for mh in article.findall(".//MeshHeadingList/MeshHeading"):
        d = mh.find("DescriptorName")
        if d is None or d.text is None:
            continue
        d_text = d.text.strip()
        d_ui = d.get("UI") or ""
        d_major = (d.get("MajorTopicYN") == "Y")

        descriptors.append(d_text)
        descriptors_ui.append(d_ui)
        if d_major:
            majors.append(d_text)

        qs = mh.findall("QualifierName")
        if qs:
            for q in qs:
                if q is None or q.text is None:
                    continue
                q_text = q.text.strip()
                q_ui = q.get("UI") or ""
                qualifiers.append(q_text)
                qualifiers_ui.append(q_ui)
                pairs.append(f"{d_text}/{q_text}")
        else:
            pairs.append(d_text)

    def dedupe(seq):
        seen = set(); out = []
        for x in seq:
            if x not in seen:
                seen.add(x); out.append(x)
        return out

    descriptors = dedupe(descriptors)
    descriptors_ui = dedupe(descriptors_ui)
    majors = dedupe(majors)
    qualifiers = dedupe(qualifiers)
    qualifiers_ui = dedupe(qualifiers_ui)
    pairs = dedupe(pairs)

    return {
        "keywords": "; ".join(descriptors) if descriptors else None,  # treat descriptors as "keywords"
        "mesh_descriptors": "; ".join(descriptors) if descriptors else None,
        "mesh_descriptors_ui": "; ".join(descriptors_ui) if descriptors_ui else None,
        "mesh_major_descriptors": "; ".join(majors) if majors else None,
        "mesh_qualifiers": "; ".join(qualifiers) if qualifiers else None,
        "mesh_qualifiers_ui": "; ".join(qualifiers_ui) if qualifiers_ui else None,
        "mesh_pairs": "; ".join(pairs) if pairs else None,
    }

def get_doi(article) -> Optional[str]:
    for aid in article.findall(".//ArticleIdList/ArticleId"):
        if aid.get("IdType") == "doi" and aid.text:
            return aid.text.strip()
    return None

def pubmed_url(n: int, prefix: str = BASELINE_PREFIX) -> str:
    # n is 1-based; left-pad to 4 digits
    return f"{BASE_URL}/pubmed{prefix}n{n:04d}.xml.gz"

def stream_rows_from_gz_url(url: str, timeout: int = 120, max_retries: int = 3, backoff_sec: float = 1.5) -> Iterable[Dict[str, Optional[str]]]:
    """Yield row dicts by streaming and parsing the gzipped XML directly from HTTP."""
    attempt = 0
    while True:
        try:
            resp = requests.get(url, stream=True, timeout=timeout)
            resp.raise_for_status()
            break
        except Exception as e:
            attempt += 1
            if attempt >= max_retries:
                raise
            time.sleep(backoff_sec * attempt)

    with gzip.GzipFile(fileobj=resp.raw, mode="rb") as gz:
        for event, elem in ET.iterparse(gz, events=("end",)):
            if elem.tag == "PubmedArticle":
                mesh = get_mesh(elem)
                row = {
                    "pmid": find_first(elem, ".//MedlineCitation/PMID"),
                    "date_published": get_pub_date(elem),
                    "date_revised": get_date_revised(elem),
                    "journal": get_journal(elem),
                    "title": get_title(elem),
                    "authors": get_authors(elem),
                    "key_substances": get_key_substances(elem),
                    "keywords": mesh["keywords"],
                    "mesh_descriptors": mesh["mesh_descriptors"],
                    "mesh_descriptors_ui": mesh["mesh_descriptors_ui"],
                    "mesh_major_descriptors": mesh["mesh_major_descriptors"],
                    "mesh_qualifiers": mesh["mesh_qualifiers"],
                    "mesh_qualifiers_ui": mesh["mesh_qualifiers_ui"],
                    "mesh_pairs": mesh["mesh_pairs"],
                    "doi": get_doi(elem),
                }
                yield row
                elem.clear()

def ingest_pubmed_range(
    start_n: int,
    end_n: int,
    out_csv_path: str,
    prefix: str = BASELINE_PREFIX,
    flush_every: int = 1000
) -> List[int]:
    """
    Stream-parse pubmed baseline files pubmed{prefix}n000{start..end}.xml.gz
    and write a single CSV incrementally.

    Returns a list of file indices that failed.
    """
    error_files: List[int] = []
    total_rows = 0

    # open once and append rows as we go
    header_written = False
    with open(out_csv_path, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=CSV_COLUMNS)
        writer.writeheader()
        header_written = True

        for n in range(start_n, end_n + 1):
            url = pubmed_url(n, prefix=prefix)
            try:
                print(f"[INFO] Processing {n:04d} → {url}")
                buffered_rows = []
                for row in stream_rows_from_gz_url(url):
                    # Ensure all expected keys exist for DictWriter
                    buffered_rows.append({k: row.get(k) for k in CSV_COLUMNS})
                    # Periodically flush to disk
                    if len(buffered_rows) >= flush_every:
                        writer.writerows(buffered_rows)
                        f.flush()
                        total_rows += len(buffered_rows)
                        buffered_rows.clear()

                # flush remainder
                if buffered_rows:
                    writer.writerows(buffered_rows)
                    f.flush()
                    total_rows += len(buffered_rows)

            except Exception as e:
                print(f"[ERROR] File index {n:04d} failed: {e}", file=sys.stderr)
                error_files.append(n)
                # continue to next file

    print(f"[DONE] Wrote ~{total_rows} rows to {out_csv_path}. Errors: {error_files}")
    return error_files

# ==============================
# Example usage
# ==============================
# errors = ingest_pubmed_range(start_n=1, end_n=3, out_csv_path="pubmed_baseline_25_0001_0003.csv")
# print("Error files:", errors)
```

```python
# pip install biopython pandas lxml
from Bio import Entrez
import pandas as pd
import math, time, re
import xml.etree.ElementTree as ET

# ---------- Configure Entrez ----------
Entrez.email = "your_email@example.com"      # REQUIRED
Entrez.api_key = None                        # e.g., "YOUR_NCBI_API_KEY"
Entrez.tool = "pubmed-author-biopython-xml"

# ---------- Helpers ----------
YEAR_RE = re.compile(r"\b(19|20)\d{2}\b")
DOI_RE  = re.compile(r"\b10\.\d{4,9}/\S+\b", re.IGNORECASE)

def build_author_term(author_name, affiliation=None, orcid=None):
    parts = [f"({author_name}[Author])"]
    if affiliation:
        parts.append(f"({affiliation}[Affiliation])")
    if orcid:
        parts.append(f"(\"{orcid}\"[auid])")
    return " AND ".join(parts)

def esearch_author(term):
    with Entrez.esearch(db="pubmed", term=term, usehistory="y",
                        retmode="xml", retmax=0) as h:
        rec = Entrez.read(h)
    return int(rec["Count"]), rec["WebEnv"], rec["QueryKey"]

def norm_name(last, fore=None, initials=None):
    last = (last or "").strip().lower()
    fore = (fore or "").strip().lower()
    initials = (initials or "").strip().lower()
    # Make a compact key: "lastname|firstinitials"
    fi = (fore[:1] if fore else (initials[:1] if initials else ""))
    return f"{last}|{fi}"

def parse_pubdate(article):
    # Prefer Year element; fallback to MedlineDate string
    y = article.findtext(".//JournalIssue/PubDate/Year")
    if not y:
        md = article.findtext(".//JournalIssue/PubDate/MedlineDate")
        if md:
            m = YEAR_RE.search(md)
            y = m.group(0) if m else None
    # Build a human-readable pubdate
    parts = []
    for tag in ("Year", "Month", "Day"):
        v = article.findtext(f".//JournalIssue/PubDate/{tag}")
        if v: parts.append(v)
    pubdate = "-".join(parts) if parts else article.findtext(".//JournalIssue/PubDate/MedlineDate")
    return y, pubdate

def extract_doi(article):
    for aid in article.findall(".//ArticleIdList/ArticleId"):
        if (aid.get("IdType") or "").lower() == "doi":
            return (aid.text or "").strip()
    # fallback: scan for DOI in LID/SO if present (rare in XML path)
    return None

def extract_mesh_terms(cit):
    terms = []
    for mh in cit.findall(".//MeshHeadingList/MeshHeading/DescriptorName"):
        if mh.text:
            terms.append(mh.text.strip())
    return terms

def extract_authors(article):
    authors = []
    # Corporate authors can appear as CollectiveName (no Last/Fore)
    for a in article.findall(".//AuthorList/Author"):
        last = a.findtext("LastName")
        fore = a.findtext("ForeName")
        initials = a.findtext("Initials")
        coll = a.findtext("CollectiveName")
        # ORCID appears as Identifier with Source="ORCID"
        orcid = None
        for ident in a.findall("Identifier"):
            if (ident.get("Source") or "").upper() == "ORCID":
                orcid = (ident.text or "").strip()

        affs = []
        for aff in a.findall("AffiliationInfo/Affiliation"):
            if aff.text:
                affs.append(aff.text.strip())

        if coll and not (last or fore):
            full = coll.strip()
            authors.append({
                "full": full, "last": None, "fore": None, "initials": None,
                "orcid": None, "affiliations": affs, "key": full.lower()
            })
        else:
            parts = []
            if fore: parts.append(fore.strip())
            if last: parts.append(last.strip())
            full = " ".join(parts) if parts else (initials or "").strip()
            authors.append({
                "full": full or None,
                "last": last or None,
                "fore": fore or None,
                "initials": initials or None,
                "orcid": orcid,
                "affiliations": affs,
                "key": norm_name(last, fore, initials)
            })
    return authors

def is_target_author(author_entry, targets):
    """Return True if author_entry matches any normalized target in `targets`."""
    k = author_entry.get("key")
    if not k:  # corporate authors: never treat as the target person
        return False
    return k in targets

def normalize_targets(author_name):
    """
    Derive a set of possible normalized name keys from the query, e.g.
    'Ng AY' -> {'ng|a'} ; 'Smith, John A' -> {'smith|j'}
    """
    name = author_name.replace(",", " ").strip()
    toks = [t for t in name.split() if t]
    if not toks:
        return set()
    # Heuristic: last token is last name if comma not present; if comma existed, first token is last
    # We'll just include both possibilities to be safe
    last_candidates = {toks[-1].lower(), toks[0].lower()}
    first_initial = toks[0][0].lower() if len(toks) > 1 else ""
    # Also look for any token of length 1 as initial
    initials = [t[0].lower() for t in toks if len(t) == 1]
    fi_candidates = {first_initial} | set(initials)
    out = set()
    for ln in last_candidates:
        for fi in fi_candidates:
            out.add(f"{ln}|{fi}")
    return out

def fetch_author_pubs_with_mesh_and_coauthors(author_name, affiliation=None, orcid=None,
                                              throttle=0.34, retmax=200):
    """
    Returns DataFrame with:
      pmid, title, journal, pubdate, year, doi, mesh_terms (list),
      authors (list of dicts), coauthors (list of dicts excluding target)
    """
    term = build_author_term(author_name, affiliation=affiliation, orcid=orcid)
    count, webenv, query_key = esearch_author(term)
    if count == 0:
        cols = ["pmid","title","journal","pubdate","year","doi","mesh_terms","authors","coauthors"]
        return pd.DataFrame(columns=cols)

    target_keys = normalize_targets(author_name)
    rows = []
    pages = math.ceil(count / retmax)

    for i in range(pages):
        retstart = i * retmax
        with Entrez.efetch(db="pubmed", query_key=query_key, WebEnv=webenv,
                           retstart=retstart, retmax=retmax,
                           rettype="full", retmode="xml") as h:
            xml_bytes = h.read()
        root = ET.fromstring(xml_bytes)  # PubmedArticleSet

        for art in root.findall(".//PubmedArticle"):
            cit = art.find("./MedlineCitation")
            article = art.find("./MedlineCitation/Article")
            if article is None:
                continue

            pmid = (cit.findtext("./PMID") or "").strip()
            title = (article.findtext("./ArticleTitle") or "").strip() or None
            journal = (article.findtext("./Journal/Title") or
                       article.findtext("./Journal/ISOAbbreviation") or None)
            year, pubdate = parse_pubdate(article)
            doi = extract_doi(art)
            mesh_terms = extract_mesh_terms(cit)

            authors = extract_authors(article)

            rows.append({
                "pmid": pmid,
                "title": title,
                "journal": journal,
                "pubdate": pubdate,
                "year": year,
                "doi": doi,
                "mesh_terms": mesh_terms,
                "authors": authors,
            })
        time.sleep(throttle)

    df = pd.DataFrame(rows)
    # Sort newest first
    df["year_num"] = pd.to_numeric(df["year"], errors="coerce")
    df = df.sort_values(["year_num", "pmid"], ascending=[False, False]).drop(columns=["year_num"])
    return df

```
